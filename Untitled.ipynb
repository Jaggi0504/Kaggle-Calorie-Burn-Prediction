```python
def detecting_outliers(val):
    sns.boxplot(x=f'{val}', data=train_df)
    plt.title("Box plot")
    plt.show()
```

```python
detecting_outliers('Age')
detecting_outliers('Height')
detecting_outliers('Weight')
detecting_outliers('Duration')
detecting_outliers('Heart_Rate')
detecting_outliers('Body_Temp')
```

```python
def get_iqr_bounds(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return lower_bound, upper_bound

def cap_outliers_with_bounds(df, column, lower_bound, upper_bound):
    df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
    return df

numerical_cols = ['Age', 'Height', 'Weight', 'Duration', 'Heart_Rate', 'Body_Temp']
for col in numerical_cols:
    lower, upper = get_iqr_bounds(train_df, col)
    train_df = cap_outliers_with_bounds(train_df, col, lower, upper)
    test_df = cap_outliers_with_bounds(test_df, col, lower, upper)
```

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import Ridge
```

```python
N_SPLITS = 5
kf = KFold(n_splits=N_SPLITS, shuffle=True, random_state=42)

oof_cat = np.zeros(len(X1_train_scaled))
oof_lgb = np.zeros(len(X1_train_scaled))
oof_xgb = np.zeros(len(X1_train_scaled))

ridge_targets = np.zeros(len(X1_train_scaled))

ridge_test_features = np.column_stack([
    cat_model.predict(test_df1_scaled),
    lgb_model.predict(test_df1_scaled),
    xgb_model.predict(test_df1_scaled)
])
```

```python
for train_idx, val_idx in kf.split(X1_train_scaled):
    X_tr, X_val = X1_train_scaled[train_idx], X1_train_scaled[val_idx]
    y_tr, y_val = y1_train_log.iloc[train_idx], y1_train_log.iloc[val_idx]
    
    cat = CatBoostRegressor(
        learning_rate=0.005,
        n_estimators=8192,
        early_stopping_rounds=11,
        verbose=0,
        random_seed=42
    )
    cat.fit(X_tr, y_tr, eval_set=(X_val, y_val))
    oof_cat[val_idx] = cat.predict(X_val)
    
    lgb = lgb_model  # already trained
    oof_lgb[val_idx] = lgb.predict(X_val)
    
    xgb = xgb_model  # already trained
    oof_xgb[val_idx] = xgb.predict(X_val)
    
    ridge_targets[val_idx] = y_val
```

```python
X_meta_train = np.column_stack([oof_cat, oof_lgb, oof_xgb])

ridge = Ridge(alpha=1.0)
ridge.fit(X_meta_train, ridge_targets)
```

```python
final_pred_log = ridge.predict(ridge_test_features)
final_pred_log = np.expm1(final_pred_log)
```

```python
rmsle = np.sqrt(mean_squared_log_error(test_df1_scaled, final_pred_log))
print("Stacked Ridge RMSLE:", rmsle)
```
